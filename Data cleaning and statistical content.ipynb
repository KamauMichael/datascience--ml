```python
# A function used for separation of characters from integers  
def wgi(x):     
    if type(x) != float:
        y= x.split(" ")
        z= int(y[1])
        return z     
    else:         
        return x
```

```python
t = other_movies['worlwide_gross_income'].apply(wgi)

t
```
